## MODE WISE DEPATURES

```python
def calculate_mode_multinomial_probabilities(df):
  air = df['DEPARTURE'][0]
  land = df['DEPARTURE'][1]
  sea = df['DEPARTURE'][2]
  all = air + land + sea
  probabilities_air = air/all
  probabilities_land = land/all
  probabilities_sea = sea/all
  multinomial_port_probabilities = [probabilities_air, probabilities_land, probabilities_sea]
  return multinomial_port_probabilities
```

```python
multinomial_mode_probabilities2013 = calculate_mode_multinomial_probabilities(df_depatureMode2013)
multinomial_mode_probabilities2014 = calculate_mode_multinomial_probabilities(df_depatureMode2014)
multinomial_mode_probabilities2016 = calculate_mode_multinomial_probabilities(df_depatureMode2016)
multinomial_mode_probabilities2017 = calculate_mode_multinomial_probabilities(df_depatureMode2017)
multinomial_mode_probabilities2018 = calculate_mode_multinomial_probabilities(df_depatureMode2018)

# year 2019 has for modes air, land, sea, rail ignoring rail for 2019
multinomial_mode_probabilities2019 = calculate_mode_multinomial_probabilities(df_depatureMode2019)
multinomial_mode_probabilities2020 = calculate_mode_multinomial_probabilities(df_depatureMode2020)
multinomial_mode_probabilities2021 = calculate_mode_multinomial_probabilities(df_depatureMode2021)
multinomial_mode_probabilities2022 = calculate_mode_multinomial_probabilities(df_depatureMode2022)


```

```python
import matplotlib.pyplot as plt

years = [2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
labels = ['Air', 'Land', 'Sea']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

fig, axes = plt.subplots(3, 3, figsize=(15, 15))
axes = axes.flatten()

for i, probabilities in enumerate(multinomial_mode_probabilities):
    ax = axes[i]
    ax.bar(labels, probabilities, color=colors)  # Apply the colors for each label
    ax.set_title(f'Multinomial Distribution of Departure Modes - {years[i]}')
    ax.set_ylabel('Probability')

plt.tight_layout()
plt.show()
```

```python
import math

def calculate_entropy(probabilities):
    entropy = 0
    for prob in probabilities:
        if prob > 0:
            entropy -= prob * math.log2(prob)
    return entropy

entropies = []
for probs in multinomial_mode_probabilities:
    entropies.append(calculate_entropy(probs))

# Find the year with maximum and minimum entropy
max_entropy_index = entropies.index(max(entropies))
min_entropy_index = entropies.index(min(entropies))

max_entropy_year = years[max_entropy_index]
min_entropy_year = years[min_entropy_index]

print(f"Year with maximum entropy: {max_entropy_year}, Entropy: {max(entropies)}")
print(f"Year with minimum entropy: {min_entropy_year}, Entropy: {min(entropies)}")
```

## STATE UT-WISE DOMESTIC AND FOREIGN TOURIST VISITS

```python
def calculate_state_multinomial_probabilities(visits):
  total = visits[0] + visits[1]
  foreign = visits[0]/total
  domestic = visits[1]/total
  return [foreign, domestic]
```

```python
def calculate_statewise_multinomial_probabilities(df):
  state_foreign_probabilities = []
  state_domestic_probabilities = []
  states = []
  for index, row in df.iterrows():
     prob = [row['Foreign'], row['Domestic']]
     foreign, domestic = calculate_state_multinomial_probabilities(prob)
     state_foreign_probabilities.append(foreign)
     state_domestic_probabilities.append(domestic)
     states.append(row['State/UT'])
  data = {'State': states, 'foreign': state_foreign_probabilities, 'domestic': state_domestic_probabilities}
  df = pd.DataFrame(data)
  return df
```

```python
df2013= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2013)
df2014= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2014)
df2016= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2016)
df2017= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2017)
df2018= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2018)
df2019= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2019)
df2020= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2020)
df2021= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2021)
df2022= calculate_statewise_multinomial_probabilities(df_stForeignDomestic2022)
```

```python
dfs = [df2013, df2014, df2016, df2017, df2018, df2019, df2020, df2021, df2022]
titles = [
    'Multinomial Distribution of Tourists in 2013',
    'Multinomial Distribution of Tourists in 2014',
    'Multinomial Distribution of Tourists in 2016',
    'Multinomial Distribution of Tourists in 2017',
    'Multinomial Distribution of Tourists in 2018',
    'Multinomial Distribution of Tourists in 2019',
    'Multinomial Distribution of Tourists in 2020',
    'Multinomial Distribution of Tourists in 2021',
    'Multinomial Distribution of Tourists in 2022'
]

# Create subplots for each DataFrame
fig, axes = plt.subplots(5, 2, figsize=(18, 24))  # 5 rows, 2 columns for 9 DataFrames
axes = axes.flatten()  # Flatten to easily loop over

# Loop through the DataFrames and their corresponding axes
for i, df in enumerate(dfs):
    ax = axes[i]
    # Create side-by-side bars for 'foreign' and 'domestic'
    width = 0.35  # Set the width of the bars
    x = range(len(df))  # Positions on the x-axis for the bars

    ax.bar([i - width / 2 for i in x], df['foreign'], width, label='Foreign', color='skyblue')
    ax.bar([i + width / 2 for i in x], df['domestic'], width, label='Domestic', color='orange')

    # Set the x-axis tick labels to state names
    ax.set_xticks(x)
    ax.set_xticklabels(df['State'], rotation=45, ha='right')

    # Set the y-axis label and title
    ax.set_ylabel('Proportion of Tourists')
    ax.set_title(titles[i])

    # Add a legend
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# prompt: calcluate state wise entropies for dfs

def calculate_state_entropy(df):
    state_entropies = {}
    for index, row in df.iterrows():
        probabilities = [row['foreign'], row['domestic']]
        entropy = calculate_entropy(probabilities)
        state_entropies[row['State']] = entropy
    return state_entropies

statewise_entropies = []
for i, df in enumerate(dfs):
    statewise_entropies.append(calculate_state_entropy(df))

# Print or further proc byess the statewise_entropies
for i, entropies in enumerate(statewise_entropies):
    print(f"State-wise entropies for {years[i]}:")
    for state, entropy in entropies.items():
        print(f"{state}: {entropy}")
    print("-" * 60)
```
