# Required Libraries for This Notebook

To run this notebook successfully, the following Python libraries are required:

1. **Pandas**
   - **Pandas** is a data manipulation and analysis library. It provides data structures like `DataFrame`, which makes handling structured data much easier.
   - To install Pandas, run:
     ```bash
     pip install pandas
     ```

2. **Scikit-learn**
   - **Scikit-learn** is a library for machine learning. It provides simple tools for data mining and data analysis, including algorithms for classification, regression, clustering, and dimensionality reduction.
   - To install Scikit-learn, run:
     ```bash
     pip install scikit-learn
     ```

3. **NumPy**
   - **NumPy** is a core scientific computing library. It supports large, multi-dimensional arrays and matrices, along with a collection of mathematical functions to operate on these arrays.
   - To install NumPy, run:
     ```bash
     pip install numpy
     ```

4. **Matplotlib**
   - **Matplotlib** is a plotting library used for generating various types of charts and visualizations, including line plots, bar charts, and histograms.
   - To install Matplotlib, run:
     ```bash
     pip install matplotlib
     ```
