## DIFFERENCES IN DEPATURES BETWEEN CONSECUTIVE YEARS

```python
df2013 = pd.read_excel(QuarterWiseIndianNationalsDepatures2013)
df2014 = pd.read_excel(QuarterWiseIndianNationalsDepatures2014)
df2016 = pd.read_excel(QuarterWiseIndianNationalsDepatures2016)
df2017 = pd.read_excel(QuarterWiseIndianNationalsDepatures2017)
df2018 = pd.read_excel(QuarterWiseIndianNationalsDepatures2018)
df2019 = pd.read_excel(QuarterWiseIndianNationalsDepatures2019)
df2020 = pd.read_excel(QuarterWiseIndianNationalsDepatures2020)
df2021 = pd.read_excel(QuarterWiseIndianNationalsDepatures2021)
df2022 = pd.read_excel(QuarterWiseIndianNationalsDepatures2022)
```

```python
Depatures2013 = df2013['Departurtres'][0]
Depatures2014 = df2014['Departurtres'][0]
Depatures2016 = df2016['Departurtres'][0]
Depatures2017 = df2017['Departurtres'][0]
Depatures2018 = df2018['Departurtres'][0]
Depatures2019 = df2019['Departurtres'][0]
Depatures2020 = df2020['Departurtres'][0]
Depatures2021 = df2021['Departurtres'][0]
Depatures2022 = df2022['Departurtres'][0]
```

```python
depatures = [Depatures2013, Depatures2014, Depatures2016, Depatures2017, Depatures2018, Depatures2019, Depatures2020, Depatures2021, Depatures2022]
```

```python
depatures_differences = [depatures[i+1] - depatures[i] for i in range(len(depatures)-1)]
```

```python
years = [2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
```

```python
import numpy as np
import matplotlib.pyplot as plt


years_labels = []
for i in range(1,len(years)):
  if i == 0:
    years_labels.append(str(years[i]) + "-" + str(years[i-1]))
  else:
    years_labels.append(str(years[i]) + "-" + str(years[i-1]))

plt.figure(figsize=(15, 6))
bars = plt.bar(years_labels, depatures_differences, color=['green' if diff >=0 else 'red' for diff in depatures_differences ])

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Difference in Depatures( IN Crores )")
plt.title("Difference in INDIAN NATIONALS DEPATURES per Year")


for bar, difference in zip(bars, depatures_differences):
    yval = bar.get_height()
    position = yval + 0.1 if difference >= 0 else yval - 0.1
    plt.text(bar.get_x() + bar.get_width()/2, position, round(abs(difference), 2), ha='center', va='bottom' if difference >= 0 else 'top')


plt.show()
```

```python
percentages = [(depatures_differences[i]/depatures[i]*100) for i in range(len(depatures)-1)]
rounded_percentages = [round(p, 2) for p in percentages]
rounded_percentages
```

```python
plt.figure(figsize=(15, 6))
bars = plt.bar(years_labels, rounded_percentages, color=['green' if p >=0 else 'red' for p in rounded_percentages])

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Percentage Difference in Departures")
plt.title("Percentage Difference in INDIAN NATIONALS DEPARTURES per Year")

for bar, percentage in zip(bars, rounded_percentages):
    yval = bar.get_height()
    position = yval + 0.1 if percentage >= 0 else yval - 0.1
    plt.text(bar.get_x() + bar.get_width()/2, position, f"{abs(percentage)}%", ha='center', va='bottom' if percentage >= 0 else 'top')

plt.show()
```

## DIFFERENCES IN DEPATURES BETWEEN MONTHS OF CONSECUTIVE YEARS

```python
df2013 = pd.read_excel(MonthWiseDepatures2013)
df2014 = pd.read_excel(MonthWiseDepatures2014)
df2016 = pd.read_excel(MonthWiseDepatures2016)
df2017 = pd.read_excel(MonthWiseDepatures2017)
df2018 = pd.read_excel(MonthWiseDepatures2018)
df2019 = pd.read_excel(MonthWiseDepatures2019)
df2020 = pd.read_excel(MonthWiseDepatures2020)
df2021 = pd.read_excel(MonthWiseDepatures2021)
df2022 = pd.read_excel(MonthWiseDepatures2022)
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
import matplotlib.pyplot as plt


def plot_monthly_differences(year1, year2, df_year1, df_year2):

    month_diffs = df_year2['Indian Nationals Departures (Nos.)'] - df_year1['Indian Nationals Departures (Nos.)']

    plt.figure(figsize=(15, 6))
    bars = plt.bar(months, month_diffs, color=['green' if diff >= 0 else 'red' for diff in month_diffs])

    plt.xlabel("Month")
    plt.ylabel("Difference in Arrivals( Millions )")
    plt.title(f"Difference in Monthly Tourist Arrivals between {year2} and {year1}")

    for bar, difference in zip(bars, month_diffs):
        yval = bar.get_height()
        position = yval + 0.1 if difference >= 0 else yval - 0.1
        plt.text(bar.get_x() + bar.get_width()/2, position, round(abs(difference), 2), ha='center', va='bottom' if difference >= 0 else 'top')

    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()


# Example usage for 2013 and 2014:  (adapt the years and dataframe names as needed)
plot_monthly_differences(2013, 2014, df2013, df2014)
plot_monthly_differences(2014, 2015, df2014, df2016)
plot_monthly_differences(2016, 2017, df2016, df2017)
plot_monthly_differences(2017, 2018, df2017, df2018)
plot_monthly_differences(2018, 2019, df2018, df2019)
plot_monthly_differences(2019, 2020, df2019, df2020)
plot_monthly_differences(2020, 2021, df2020, df2021)
plot_monthly_differences(2021, 2022, df2021, df2022)
```

```python
def plot_monthly_percentage_differences(year1, year2, df_year1, df_year2):

    month_diffs = df_year2['Indian Nationals Departures (Nos.)'] - df_year1['Indian Nationals Departures (Nos.)']
    percentages = (month_diffs / df_year1['Indian Nationals Departures (Nos.)']) * 100
    rounded_percentages = [round(p, 2) for p in percentages]

    plt.figure(figsize=(15, 6))
    bars = plt.bar(months, rounded_percentages, color=['green' if p >= 0 else 'red' for p in rounded_percentages])

    plt.xlabel("Month")
    plt.ylabel("Percentage Difference in Departures")
    plt.title(f"Percentage Difference in Monthly Departures between {year2} and {year1}")

    for bar, percentage in zip(bars, rounded_percentages):
        yval = bar.get_height()
        position = yval + 0.1 if percentage >= 0 else yval - 0.1
        plt.text(bar.get_x() + bar.get_width()/2, position, f"{abs(percentage)}%", ha='center', va='bottom' if percentage >= 0 else 'top')

    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Example usage (adapt years and dataframes as needed):
plot_monthly_percentage_differences(2013, 2014, df2013, df2014)
plot_monthly_percentage_differences(2014, 2016, df2014, df2016)
plot_monthly_percentage_differences(2016, 2017, df2016, df2017)
plot_monthly_percentage_differences(2017, 2018, df2017, df2018)
plot_monthly_percentage_differences(2018, 2019, df2018, df2019)
plot_monthly_percentage_differences(2019, 2020, df2019, df2020)
plot_monthly_percentage_differences(2020, 2021, df2020, df2021)
plot_monthly_percentage_differences(2021, 2022, df2021, df2022)
```
