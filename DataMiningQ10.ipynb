```python
Arrivals
```

```python
YearWiseArrivalsDifferences = [Arrivals[i+1] - Arrivals[i] for i in range(len(Arrivals)-1)]
```

```python
percentages = [(YearWiseArrivalsDifferences[i]/Arrivals[i])*100 for i in range(len(Arrivals)-1)]
rounded_percentages = [round(p, 2) for p in percentages]
rounded_percentages
```

```python
years_labels = []
for i in range(1,len(years)):
  if i == 0:
    years_labels.append(str(years[i]) + "-" + str(years[i-1]))
  else:
    years_labels.append(str(years[i]) + "-" + str(years[i-1]))

plt.figure(figsize=(15, 6))
bars = plt.bar(years_labels, rounded_percentages, color=['green' if diff >=0 else 'red' for diff in YearWiseArrivalsDifferences ])

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Percentage Change in Arrivals")
plt.title("Percentage Change in Tourist Arrivals per Year")

for bar, percentage in zip(bars, rounded_percentages):
    yval = bar.get_height()
    position = yval + 0.1 if percentage >= 0 else yval - 0.1
    plt.text(bar.get_x() + bar.get_width()/2, position, f"{percentage}%", ha='center', va='bottom' if percentage >= 0 else 'top')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(years, Arrivals, label='Arrivals')
plt.xlabel('Year')
plt.ylabel('Number')
plt.title('Tourism Data')
plt.xticks(years)
plt.legend()
plt.grid(True)
plt.show()
```

<p><strong>Observations</strong></p>

<p><strong>Pre-COVID Growth (2013–2019)</strong></p>
<ul>
  <li>From 2013 to 2019, both tourist arrivals and Indian national departures showed steady growth.</li>
  <li>By 2019, India saw **10.93 million arrivals** and **26.92 million departures**, reflecting a healthy tourism and travel industry.</li>
</ul>

<p><strong>Impact of COVID-19 (2020–2021)</strong></p>
<ul>
  <li><strong>2020</strong> experienced a sharp decline in both arrivals and departures. Arrivals dropped by **75%** compared to 2019, while departures decreased by **73%**.</li>
  <li><strong>2021</strong> saw a smaller decline, with arrivals still down by **86%** compared to 2019, while departures fell by about **68%**.</li>
</ul>

<p><strong>Post-COVID Recovery (2022)</strong></p>
<ul>
  <li><strong>2022</strong> showed a partial recovery for both arrivals and departures. Arrivals rebounded to **6.44 million**, but were still only about **59%** of the 2019 level.</li>
  <li>Departures saw a more significant recovery, with **21.60 million** departures in 2022, reaching approximately **80%** of the 2019 level.</li>
  <li>The recovery in 2022 indicates that while international tourism is picking up, domestic travel (Indian nationals traveling abroad) has recovered more strongly.</li>
</ul>

<p><strong>Conclusion: Effect of COVID-19 on Tourism in India</strong></p>
<ul>
  <li>The COVID-19 pandemic caused significant disruptions to both **incoming tourism** and **Indian national travel** in 2020 and 2021.</li>
  <li>While **tourist arrivals** have seen some recovery in 2022, they remain below pre-pandemic levels, with a return to full recovery expected to take several more years.</li>
  <li>Indian nationals' travel abroad (departures) has rebounded more strongly in 2022, suggesting that domestic restrictions lifted earlier compared to international travel.</li>
  <li>The tourism industry and international travel to/from India may take several years to return to pre-COVID levels, with varying recovery rates for inbound vs. outbound travel.</li>
</ul>
