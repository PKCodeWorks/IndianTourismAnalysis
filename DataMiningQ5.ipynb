```python
def findminmax(year):
      max_year = np.max(year)
      argmax_year = np.argmax(year)
      min_year = np.min(year)
      argmin_year = np.argmin(year)
      return max_year, argmax_year, min_year, argmin_year
```

```python
continents = ['North America', 'Central & South America', 'Western Europe', 'Eastern Europe', 'Africa', 'West Asia', 'South Asia', 'South East Asia', 'East Asia', 'Australasia']
```

```python
continents = [continent.upper() for continent in continents]
```

```python
def continent_splits_ranges(df):
  indices = []
  index = 0
  for country in df['Destination Country']:
    if country.upper() in continents:
      indices.append(index)
    index += 1
  indices.append(len(df['Destination Country']))

  ranges = []
  for i in range(len(indices)-1):
      ranges.append([indices[i]+1, indices[i+1]])
  return ranges
```

```python
quarters = ['1st Quarter (Jan-Mar)', '2nd Quarter (Apr-June)', '3rd Quarter (July-Sep)', '4th Quarter (Oct-Dec)']
```

```python
def calcluate_continent_purpose_split(df, year):
      continent_split = continent_splits_ranges(df)
      print(f"YEAR : {year}")
      for split, continent in zip(continent_split, continents):
        Q1 =df.iloc[split[0]:split[1]]['1st Quarter (Jan-Mar)'].sum()
        Q2 =df.iloc[split[0]:split[1]]['2nd Quarter (Apr-Jun)'].sum()
        Q3 =df.iloc[split[0]:split[1]]['3rd Quarter (July-Sep)'].sum()
        Q4 =df.iloc[split[0]:split[1]]['4th Quarter (Oct-Dec)'].sum()
        max_year, argmax_year, min_year, argmin_year = findminmax([Q1, Q2, Q3, Q4])
        print(f"{continent} : PEAK QUARTER : {quarters[argmax_year]}, LEAN QUARTER : {quarters[argmin_year]}")
```

```python
continents = ['North America', 'Central & South America', 'Western Europe', 'Eastern Europe', 'Africa', 'West Asia', 'South Asia', 'South East Asia', 'East Asia', 'Australasia']
```

```python
def remove_continent_rows(df):

    # Convert both DataFrame column and continents list to lowercase for comparison
    mask = ~df['Destination Country'].str.upper().isin([c.upper() for c in continents])

    # Apply the mask to filter the DataFrame
    df_filtered = df[mask]

    return df_filtered
```

```python
def remove_other_rows_age(df):
    df = df[df['Destination Country'].str.upper() != 'OTHERS']
    return df
```

```python
df2014 = remove_continent_rows(df2014)
df2016 = remove_continent_rows(df2016)
df2017 = remove_continent_rows(df2017)
df2019 = remove_continent_rows(df2019)
df2020 = remove_continent_rows(df2020)
df2021 = remove_continent_rows(df2021)
df2022 = remove_continent_rows(df2022)
```

```python
def calcluate_country(df, year):
      print(f"YEAR : {year}")
      for index, row in df.iterrows():
    # Access data in each column for the current row
          country = row['Destination Country']
          q1 = row['1st Quarter (Jan-Mar)']
          q2 = row['2nd Quarter (Apr-Jun)']
          q3 = row['3rd Quarter (July-Sep)']
          q4 = row['4th Quarter (Oct-Dec)']
          max_year, argmax_year, min_year, argmin_year = findminmax([q1,q2,q3,q4])
          print(f"{country} : PEAK QUARTER : {quarters[argmax_year]}, LEAN QUARTER : {quarters[argmin_year]}")
```
