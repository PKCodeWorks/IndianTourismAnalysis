## BY CONTINENTS

```python
continents = ['North America', 'Central & South America', 'Western Europe', 'Eastern Europe', 'Africa', 'West Asia', 'South Asia', 'South East Asia', 'East Asia', 'Australasia']
```

```python
continents = [continent.upper() for continent in continents]
```

```python
def continent_arrivals_splits_ranges(df):
  indices = []
  index = 0
  for country in df['Country of Nationality']:
    if country.upper() in continents:
      indices.append(index)
    index += 1
  indices.append(len(df['Country of Nationality']))

  ranges = []
  for i in range(len(indices)-1):
      ranges.append([indices[i]+1, indices[i+1]])
  return ranges
```

```python
def calcluate_continent_arrivals_split(df, year):
      continent_split = continent_arrivals_splits_ranges(df)
      continentarrivals = []
      #print(f"YEAR : {year}")
      for split, continent in zip(continent_split, continents):
        sum = df.iloc[split[0]:split[1]]['Arrivals (in numbers)'].sum()
        continentarrivals.append(sum)
        #print(f"{continent} : {sum}")
      return continentarrivals
```

```python
continentarrivals2014 = calcluate_continent_arrivals_split(dfArrivals2014, 2014)
continentarrivals2016 = calcluate_continent_arrivals_split(dfArrivals2016, 2016)
continentarrivals2017 = calcluate_continent_arrivals_split(dfArrivals2017, 2017)
continentarrivals2019 = calcluate_continent_arrivals_split(dfArrivals2019, 2019)
continentarrivals2020 = calcluate_continent_arrivals_split(dfArrivals2020, 2020)
continentarrivals2021 = calcluate_continent_arrivals_split(dfArrivals2021, 2021)
continentarrivals2022 = calcluate_continent_arrivals_split(dfArrivals2022, 2022)
```

```python
def continent_depatures_splits_ranges(df):
  indices = []
  index = 0
  for country in df['Destination Country']:
    if country.upper() in continents:
      indices.append(index)
    index += 1
  indices.append(len(df['Destination Country']))

  ranges = []
  for i in range(len(indices)-1):
      ranges.append([indices[i]+1, indices[i+1]])
  return ranges
```

```python
def calcluate_continent_depatures_split(df, year):
      continent_split = continent_depatures_splits_ranges(df)
      #print(f"YEAR : {year}")
      continentdepatures = []
      for split, continent in zip(continent_split, continents):
        sum = df.iloc[split[0]:split[1]]['Departures (In Numbers)'].sum()
        continentdepatures.append(sum)
        #print(f"{continent} : {sum}")
      return continentdepatures
```

## TOP 5 CONTINENTS WITH HIGHEST POSITIVE AND NEGATIVE BALANCES

```python
def top_5_continents(balances, year):
    # Combine continents and balances into a list of tuples
    continent_balance_tuples = list(zip(continents, balances))

    # Sort the tuples by balance in descending order
    sorted_continent_balance_tuples = sorted(continent_balance_tuples, key=lambda x: x[1], reverse=True)

    # Get the top 5 continents with the highest positive balances
    top_5_positive = sorted_continent_balance_tuples[:5]

    # Sort the tuples by balance in ascending order
    sorted_continent_balance_tuples = sorted(continent_balance_tuples, key=lambda x: x[1])

    # Get the top 5 continents with the highest negative balances
    top_5_negative = sorted_continent_balance_tuples[:5]

    print(f"Top 5 Continents with Highest Positive Balances in {year}:")
    for continent, balance in top_5_positive:
        print(f"{continent}: {balance}")

    print(f"\nTop 5 Continents with Highest Negative Balances in {year}:")
    for continent, balance in top_5_negative:
        print(f"{continent}: {balance}")

    return top_5_positive, top_5_negative
```

```python
def top_5_positive_countries(dfArrivals_sorted, dfdepatures_sorted, year):
  print(f"Top 5 Countries with Highest Positive Balances in {year}:")
  countrybalances = dfArrivals_sorted['Arrivals (in numbers)'].values-dfdepatures_sorted['Departures (In Numbers)'].values
  print(dfArrivals_sorted.shape,dfdepatures_sorted.shape,len(countrybalances))
  data = {'Country': dfArrivals_sorted['Country of Nationality'],
          'Balance': countrybalances}

  countrybalances = pd.DataFrame(data)

  # Sort the DataFrame by balance in descending order
  countrybalances_sorted = countrybalances.sort_values(by='Balance', ascending=False)

  # Get the top 5 countries
  top_5_countries = countrybalances_sorted.head(5)

  return top_5_countries
```

## TOP 5 NEGATIVE BALANCE COUNTRIES

```python
def top_5_negative_countries(dfArrivals_sorted, dfdepatures_sorted, year):
  print(f"Top 5 Countries with Highest Positive Balances in {year}:")
  countrybalances = dfArrivals_sorted['Arrivals (in numbers)'].values-dfdepatures_sorted['Departures (In Numbers)'].values
  print(dfArrivals_sorted.shape,dfdepatures_sorted.shape,len(countrybalances))
  data = {'Country': dfArrivals_sorted['Country of Nationality'],
          'Balance': countrybalances}

  countrybalances = pd.DataFrame(data)

  # Sort the DataFrame by balance in descending order
  countrybalances_sorted = countrybalances.sort_values(by='Balance', ascending=True)

  # Get the top 5 countries
  top_5_countries = countrybalances_sorted.head(5)

  return top_5_countries
```
