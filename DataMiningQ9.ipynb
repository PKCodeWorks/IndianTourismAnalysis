```python
def continent_splits_ranges(df):
  indices = []
  index = 0
  continents = ['North America', 'Central & South America', 'Western Europe', 'Eastern Europe', 'Africa', 'West Asia', 'South Asia', 'South East Asia', 'East Asia', 'Australasia']
  continents = [continent.upper() for continent in continents]
  for country in df['Country of Nationality'].str.upper():
    if country in continents:
      indices.append(index)
    index += 1
  indices.append(len(df['Country of Nationality']))

  ranges = []
  for i in range(len(indices)-1):
      ranges.append([indices[i]+1, indices[i+1]])
  return ranges
```

```python
def calcluate_sums(df):
  bussiness = (df['Arrivals (in Numbers)']*df['Business and Professional']).dropna()
  leisure = (df['Arrivals (in Numbers)']*df['Leisure Holiday and Recreation']).dropna()
  medical = (df['Arrivals (in Numbers)']*df['Medical']).dropna()
  indian = (df['Arrivals (in Numbers)']*df['Indian Diaspora']).dropna()
  # droping student because dataset for some years doesn't contain Student Column
  #student = df['Arrivals (in Numbers)']*df['Student']
  others = (df['Arrivals (in Numbers)']*df['Others']).dropna()


  return [bussiness.sum(), leisure.sum(), medical.sum(), indian.sum(), others.sum()]
```

```python
def calcluate_continent_purpose_split(df, continent_splits, year):
      print(f"Year: {year}")
      column_names = ['Business and Professional', 'Leisure Holiday and Recreation', 'Medical', 'Indian Diaspora', 'Others']
      all_results = []

      for split in range(len(continent_splits)):
         all_results.append(calcluate_sums(df[continent_splits[split][0]:continent_splits[split][1]]))

      df_empty = pd.DataFrame(all_results, columns=column_names)
      return df_empty
```

```python
def concat_dfs(data2016, data2017, data2018, data2019, data2020, data2021, data2022, index):
    column_names = ['Business and Professional', 'Leisure Holiday and Recreation', 'Medical', 'Indian Diaspora', 'Others']

    dfs_to_concat = [
        pd.DataFrame(data2016.iloc[index]).T,
        pd.DataFrame(data2017.iloc[index]).T,
        pd.DataFrame(data2018.iloc[index]).T,
        pd.DataFrame(data2019.iloc[index]).T,
        pd.DataFrame(data2020.iloc[index]).T,
        pd.DataFrame(data2021.iloc[index]).T,
        pd.DataFrame(data2022.iloc[index]).T
    ]

    # Use pd.concat to combine the DataFrames
    df = pd.concat(dfs_to_concat, ignore_index=True)
    return df
```

```python
def linearInterpolate(df, year, index):
  # print(f"Year: {year}")
  # Linear interpolation for all features
  df_linear = df.copy()
  df_linear = df_linear.set_index('Year')  # Set 'Year' as index for interpolation
  df_linear_interpolated = df_linear.interpolate(method='linear', axis=0)

  # print("\nDataFrame after Linear Interpolation:")
  return df_linear_interpolated
```

```python
def meanInterpolate(df, year, index):
    # print(f"Year: {year}")
    # Mean imputation for all features
    df_mean = df.copy()
    for column in df_mean.columns[1:]:  # Skip the 'Year' column
        mean_value = df_mean[column].mean()
        df_mean[column] = df_mean[column].fillna(mean_value)

    # print("\nDataFrame after Mean Imputation:")
    return df_mean
```

```python
from sklearn.impute import KNNImputer

def knnInterpolate(df, year, index):
    # print(f"Year: {year}")
    # Drop the 'Year' column for imputation as it is not a feature to be used for KNN
    df_without_year = df.drop('Year', axis=1)

    # Initialize KNN Imputer with k=2 (use the 2 nearest neighbors for imputation)
    knn_imputer = KNNImputer(n_neighbors=2)

    # Apply KNN imputation
    df_imputed = pd.DataFrame(knn_imputer.fit_transform(df_without_year), columns=df_without_year.columns)

    # Add back the 'Year' column
    df_imputed['Year'] = df['Year']

    # Display the imputed DataFrame
    # print("\nDataFrame after KNN Imputation (for missing years):")
    return df_imputed
```

```python
def rmse(imputed, actual):
    return np.sqrt(((imputed - actual) ** 2).mean())
```

```python
def createdf(df, year, index):
    data = {
        'Year': [2016, 2017, 2018, 2019, 2020, 2021, 2022],
        'Business and Professional': df['Business and Professional'],
        'Leisure Holiday and Recreation': df['Leisure Holiday and Recreation'],
        'Medical': df['Medical'],
        'Indian Diaspora': df['Indian Diaspora'],
        'Others': df['Others']
    }

    df = pd.DataFrame(data)
    df.iloc[index] = [year, np.nan, np.nan, np.nan, np.nan, np.nan,]
    return df
```

```python
def continent_splits_ranges(df):
  indices = []
  index = 0
  continents = ['North America', 'Central & South America', 'Western Europe', 'Eastern Europe', 'Africa', 'West Asia', 'South Asia', 'South East Asia', 'East Asia', 'Australasia']
  continents = [continent.upper() for continent in continents]
  for country in df['Country of Nationality'].str.upper():
    if country in continents:
      indices.append(index)
    index += 1
  indices.append(len(df['Country of Nationality']))

  ranges = []
  for i in range(len(indices)-1):
      ranges.append([indices[i]+1, indices[i+1]])
  return ranges
```

```python
def concat_dfs(data2013, data2014, data2016, data2017, data2018, data2019, data2020, data2021, data2022, index):
    column_names = ["Arrivals (in numbers)"]

    dfs_to_concat = [
        pd.DataFrame(data2013.iloc[index]).T,
        pd.DataFrame(data2014.iloc[index]).T,
        pd.DataFrame(data2016.iloc[index]).T,
        pd.DataFrame(data2017.iloc[index]).T,
        pd.DataFrame(data2018.iloc[index]).T,
        pd.DataFrame(data2019.iloc[index]).T,
        pd.DataFrame(data2020.iloc[index]).T,
        pd.DataFrame(data2021.iloc[index]).T,
        pd.DataFrame(data2022.iloc[index]).T
    ]

    # Use pd.concat to combine the DataFrames
    df = pd.concat(dfs_to_concat, ignore_index=True)
    return df
```

```python
def linearInterpolate(df, year, index):
  # print(f"Year: {year}")
  # Linear interpolation for all features
  df_linear = df.copy()
  df_linear = df_linear.set_index('Year')  # Set 'Year' as index for interpolation
  df_linear_interpolated = df_linear.interpolate(method='linear', axis=0)
  return df_linear_interpolated
```

```python
def meanInterpolate(df, year, index):
    # print(f"Year: {year}")
    # Mean imputation for all features
    df_mean = df.copy()
    for column in df_mean.columns[1:]:  # Skip the 'Year' column
        mean_value = df_mean[column].mean()
        df_mean[column] = df_mean[column].fillna(mean_value)
    return df_mean
```

```python
from sklearn.impute import KNNImputer

def knnInterpolate(df, year, index):
    # print(f"Year: {year}")
    # Drop the 'Year' column for imputation as it is not a feature to be used for KNN
    df_without_year = df.drop('Year', axis=1)

    # Initialize KNN Imputer with k=2 (use the 2 nearest neighbors for imputation)
    knn_imputer = KNNImputer(n_neighbors=2)

    # Apply KNN imputation
    df_imputed = pd.DataFrame(knn_imputer.fit_transform(df_without_year), columns=df_without_year.columns)

    # Add back the 'Year' column
    df_imputed['Year'] = df['Year']

    return df_imputed
```

```python
def rmse(imputed, actual):
    return np.sqrt(((imputed - actual) ** 2).mean())
```

```python
def createQdf(df, year, index):
    data = {
        'Year': [2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
        'Arrivals (in numbers)': df['Arrivals (in numbers)'],
    }

    df = pd.DataFrame(data)
    df.iloc[index] = [year, np.nan]
    return df
```

## Explain the differences between the techniques.

<h1>Differences Between KNN Imputation, Mean Imputation, and Linear Interpolation</h1>

<h2>1. K-Nearest Neighbors (KNN) Imputation</h2>

<h3>How it works:</h3>
<p><strong>KNN imputation</strong> fills in missing data by using the values of "neighbors," which are the data points that are most similar to the missing point.</p>
<p>The "similarity" between data points is typically measured using <strong>Euclidean distance</strong> or other distance metrics.</p>
<p>For each missing value, the algorithm identifies the <strong>K nearest neighbors</strong> (based on the available data points) and imputes the missing value as the <strong>average (or weighted average)</strong> of those neighbors' values.</p>

<p><strong>Example:</strong><br>
If you are missing data for 2020, KNN would look at the values for 2019 and 2021 and impute 2020 as the average of these two values.</p>

<h3>Advantages:</h3>
<ul>
  <li><strong>Takes relationships into account</strong>: KNN finds and uses the most similar data points, reflecting the underlying data structure more accurately than simpler methods like mean imputation.</li>
  <li><strong>Works well with small datasets</strong>: Particularly useful when the missing data is sparse and neighboring points provide valuable information.</li>
</ul>

<h3>Disadvantages:</h3>
<ul>
  <li><strong>Computationally expensive</strong>: Identifying the nearest neighbors can be slow, especially with large datasets.</li>
  <li><strong>Sensitive to irrelevant features</strong>: If irrelevant features are present, KNN may incorrectly identify neighbors.</li>
  <li><strong>Struggles with sparse data</strong>: If the dataset is sparse, it may not find good neighbors to make accurate predictions.</li>
</ul>

---

<h2>2. Mean Imputation</h2>

<h3>How it works:</h3>
<p><strong>Mean imputation</strong> replaces missing values with the <strong>mean</strong> (average) of all the available values in the dataset.</p>
<p>For time-series data, it calculates the mean of all known data points and uses that mean to fill in missing years.</p>

<p><strong>Example:</strong><br>
If you're missing data for 2020, the algorithm would replace the missing value with the mean of all known values (e.g., the mean of values from 2010, 2011, 2012, etc.).</p>

<h3>Advantages:</h3>
<ul>
  <li><strong>Simple and fast</strong>: Very easy to implement and computationally efficient.</li>
  <li><strong>Preserves dataset size</strong>: No data points are discarded, which maintains the dataset's integrity.</li>
</ul>

<h3>Disadvantages:</h3>
<ul>
  <li><strong>Ignores patterns in data</strong>: Mean imputation doesn’t account for trends or temporal relationships in the data, which is problematic for time-series data.</li>
  <li><strong>Can reduce variability</strong>: Replacing missing values with the same constant value can artificially lower the natural variation in the dataset.</li>
  <li><strong>Introduces bias</strong>: In skewed datasets, mean imputation can distort the distribution and lead to biased results.</li>
</ul>

---

<h2>3. Linear Interpolation</h2>

<h3>How it works:</h3>
<p><strong>Linear interpolation</strong> estimates missing values by drawing a straight line between two known neighboring points and filling in the missing value along that line.</p>
<p>The formula for linear interpolation between two points (x1, y1) and (x2, y2) is:</p>

<pre>
y = y1 + ( (y2 - y1) / (x2 - x1) ) * (x - x1)
</pre>

<p>Where:
  <br>x1 and x2 are the years before and after the missing value.
  <br>y1 and y2 are the known values for those years.
  <br>x is the year for which we want to interpolate.</p>

<p><strong>Example:</strong><br>
If you're missing data for 2020, linear interpolation would calculate the value by assuming a linear change between the values for 2019 and 2021.</p>

<h3>Advantages:</h3>
<ul>
  <li><strong>Simple and intuitive</strong>: Easy to understand and implement.</li>
  <li><strong>Preserves trends</strong>: It’s effective when the data changes linearly over time, especially in time-series data.</li>
</ul>

<h3>Disadvantages:</h3>
<ul>
  <li><strong>Assumes linearity</strong>: This method works only when data changes in a linear fashion. If the data has non-linear patterns, it might produce inaccurate imputations.</li>
  <li><strong>Sensitive to outliers</strong>: Extreme values in neighboring points can distort the interpolation, leading to biased estimates.</li>
</ul>

<h2>Summary of When to Use Each Technique</h2>

<ul>
  <li><strong>KNN Imputation</strong> is best when:
    <ul>
      <li>You have a <strong>small dataset</strong> with well-defined relationships between data points.</li>
      <li>The missing values can be predicted based on the similarity of other data points (i.e., you have other features or years that are similar).</li>
    </ul>
  </li>

  <li><strong>Mean Imputation</strong> is best when:
    <ul>
      <li>You need a <strong>quick and simple</strong> solution.</li>
      <li>The missing values are relatively <strong>small in number</strong>.</li>
      <li>You don't expect any major trends or patterns in the data that could be captured by more complex methods.</li>
    </ul>
  </li>

  <li><strong>Linear Interpolation</strong> is best when:
    <ul>
      <li>The data is from a <strong>time-series</strong> and you believe the values between time points change <strong>linearly</strong>.</li>
      <li>You have <strong>short gaps</strong> in the data and expect a gradual change in values over time.</li>
    </ul>
  </li>
</ul>
