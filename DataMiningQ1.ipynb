## DIFFERENCES IN FTAs ARRIVALS BETWEEN CONSECUTIVE YEARS

```python
df2013 = pd.read_excel(FTAsbyQuarters2013)
df2014 = pd.read_excel(FTAsbyQuarters2014)
df2016 = pd.read_excel(FTAsbyQuarters2016)
df2017 = pd.read_excel(FTAsbyQuarters2017)
df2018 = pd.read_excel(FTAsbyQuarters2018)
df2019 = pd.read_excel(FTAsbyQuarters2019)
df2020 = pd.read_excel(FTAsbyQuarters2020)
df2021 = pd.read_excel(FTAsbyQuarters2021)
df2022 = pd.read_excel(FTAsbyQuarters2022)
```

```python
Arrivals2013 = df2013['ARRIVALS'][0]
Arrivals2014 = df2014['ARRIVALS'][0]
Arrivals2016 = df2016['ARRIVALS'][0]
Arrivals2017 = df2017['ARRIVALS'][0]
Arrivals2018 = df2018['ARRIVALS'][0]
Arrivals2019 = df2019['ARRIVALS'][0]
Arrivals2020 = df2020['ARRIVALS'][0]
Arrivals2021 = df2021['ARRIVALS'][0]
Arrivals2022 = df2022['ARRIVALS'][0]
```

```python
Arrivals = [Arrivals2013, Arrivals2014, Arrivals2016, Arrivals2017, Arrivals2018, Arrivals2019, Arrivals2020, Arrivals2021, Arrivals2022]
```

```python
YearWiseArrivalsDifferences = [Arrivals[i+1] - Arrivals[i] for i in range(len(Arrivals)-1)]
```

```python
years = [2013, 2014, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
```

```python
YearWiseArrivalsDifferences
```

```python
import numpy as np
import matplotlib.pyplot as plt


years_labels = []
for i in range(1,len(years)):
  if i == 0:
    years_labels.append(str(years[i]) + "-" + str(years[i-1]))
  else:
    years_labels.append(str(years[i]) + "-" + str(years[i-1]))

plt.figure(figsize=(15, 6))
bars = plt.bar(years_labels, YearWiseArrivalsDifferences, color=['green' if diff >=0 else 'red' for diff in YearWiseArrivalsDifferences ])

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Difference in Arrivals( IN Millions )")
plt.title("Difference in Tourist Arrivals per Year")


for bar, difference in zip(bars, YearWiseArrivalsDifferences):
    yval = bar.get_height()
    position = yval + 0.1 if difference >= 0 else yval - 0.1
    plt.text(bar.get_x() + bar.get_width()/2, position, round(abs(difference), 2), ha='center', va='bottom' if difference >= 0 else 'top')


plt.show()
```

```python
percentages = [(YearWiseArrivalsDifferences[i]/Arrivals[i])*100 for i in range(len(Arrivals)-1)]
rounded_percentages = [round(p, 2) for p in percentages]
rounded_percentages
```

```python
plt.figure(figsize=(15, 6))
bars = plt.bar(years_labels, rounded_percentages, color=['green' if diff >=0 else 'red' for diff in YearWiseArrivalsDifferences ])

# Add labels and title
plt.xlabel("Year")
plt.ylabel("Percentage Change in Arrivals")
plt.title("Percentage Change in Tourist Arrivals per Year")

for bar, percentage in zip(bars, rounded_percentages):
    yval = bar.get_height()
    position = yval + 0.1 if percentage >= 0 else yval - 0.1
    plt.text(bar.get_x() + bar.get_width()/2, position, f"{percentage}%", ha='center', va='bottom' if percentage >= 0 else 'top')

plt.show()
```

## DIFFERENCES IN FTAs ARRIVALS BETWEEN MONTHS OF CONSECUTIVE YEARS

```python
df2013 = pd.read_excel(MonthWiseFTAs2013)
df2014 = pd.read_excel(MonthWiseFTAs2014)
df2016 = pd.read_excel(MonthWiseFTAs2016)
df2017 = pd.read_excel(MonthWiseFTAs2017)
df2018 = pd.read_excel(MonthWiseFTAs2018)
df2019 = pd.read_excel(MonthWiseFTAs2019)
df2020 = pd.read_excel(MonthWiseFTAs2020)
df2021 = pd.read_excel(MonthWiseFTAs2021)
df2022 = pd.read_excel(MonthWiseFTAs2022)
```

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
```

```python
import matplotlib.pyplot as plt


def plot_monthly_differences(year1, year2, df_year1, df_year2):

    month_diffs = df_year2['Foreign tourist arrivals'] - df_year1['Foreign tourist arrivals']

    plt.figure(figsize=(15, 6))
    bars = plt.bar(months, month_diffs, color=['green' if diff >= 0 else 'red' for diff in month_diffs])

    plt.xlabel("Month")
    plt.ylabel("Difference in Arrivals( Millions )")
    plt.title(f"Difference in Monthly Tourist Arrivals between {year2} and {year1}")

    for bar, difference in zip(bars, month_diffs):
        yval = bar.get_height()
        position = yval + 0.1 if difference >= 0 else yval - 0.1
        plt.text(bar.get_x() + bar.get_width()/2, position, round(abs(difference), 2), ha='center', va='bottom' if difference >= 0 else 'top')

    plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
    plt.tight_layout() # Adjust layout to prevent labels from overlapping
    plt.show()


# Example usage for 2013 and 2014:  (adapt the years and dataframe names as needed)
plot_monthly_differences(2013, 2014, df2013, df2014)
plot_monthly_differences(2014, 2015, df2014, df2016)
plot_monthly_differences(2016, 2017, df2016, df2017)
plot_monthly_differences(2017, 2018, df2017, df2018)
plot_monthly_differences(2018, 2019, df2018, df2019)
plot_monthly_differences(2019, 2020, df2019, df2020)
plot_monthly_differences(2020, 2021, df2020, df2021)
plot_monthly_differences(2021, 2022, df2021, df2022)
```

```python
# prompt: add percentage changes monthwise

def plot_monthly_percentage_changes(year1, year2, df_year1, df_year2):
    month_diffs = df_year2['Foreign tourist arrivals'] - df_year1['Foreign tourist arrivals']
    percentage_changes = (month_diffs / df_year1['Foreign tourist arrivals']) * 100

    plt.figure(figsize=(15, 6))
    bars = plt.bar(months, percentage_changes, color=['green' if diff >= 0 else 'red' for diff in month_diffs])

    plt.xlabel("Month")
    plt.ylabel("Percentage Change in Arrivals")
    plt.title(f"Percentage Change in Monthly Tourist Arrivals between {year2} and {year1}")

    for bar, percentage in zip(bars, percentage_changes):
        yval = bar.get_height()
        position = yval + 0.1 if percentage >= 0 else yval - 0.1
        plt.text(bar.get_x() + bar.get_width()/2, position, f"{percentage:.2f}%", ha='center', va='bottom' if percentage >= 0 else 'top')

    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Example usage
plot_monthly_percentage_changes(2013, 2014, df2013, df2014)
plot_monthly_percentage_changes(2014, 2016, df2014, df2016) #Corrected year
plot_monthly_percentage_changes(2016, 2017, df2016, df2017)
plot_monthly_percentage_changes(2017, 2018, df2017, df2018)
plot_monthly_percentage_changes(2018, 2019, df2018, df2019)
plot_monthly_percentage_changes(2019, 2020, df2019, df2020)
plot_monthly_percentage_changes(2020, 2021, df2020, df2021)
plot_monthly_percentage_changes(2021, 2022, df2021, df2022)
```
